```python
$map$

- before applying map method first put the unique labels of the column

- for examples case_status is a categorical column

    - denied
    
    - Certified
    
- create a dictionary key as label, value as number

- this dictionary 
```

```python
visa_df['case_status'].unique()
```

```python
d={'Certified':0,'Denied':1}
visa_df['case_status']=visa_df['case_status'].map(d)
visa_df.head(3)
```

```python
visa_df['continent'].unique()
```

```python
d1={"Asia":0,"Africa":1,"North America":2,"Europe":3,"South America":4,"Oceania":5}
visa_df['continent']=visa_df['continent'].map(d1)
visa_df
```

```python
d={}
labels=visa_df['continent'].unique()
for i in range(len(labels)):
    d[labels[i]]=i
visa_df['continent']=visa_df['continent'].map(d)
visa_df
```

```python
visa_df
```

```python
cat_cols=visa_df.select_dtypes(include='object').columns
cat_cols

for j in cat_cols[1:]:
    d={}
    labels=visa_df[j].unique()
    for i in range(len(labels)):
        d[labels[i]]=i
        visa_df[j]=visa_df[j].map(d)
visa_df
```
